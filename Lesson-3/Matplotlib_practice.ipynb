## 基本画图

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib notebook
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax1.set(
    xlim=[0.5, 4.5], ylim=[-2, 8], 
    title='An example', 
    xlabel='X-Axis', ylabel='Y-Axis'
      )
```

```python
fig2 = plt.figure()
ax = fig2.add_subplot(111)
ax.plot([1, 2, 3, 4], [1, 4, 9, 16], color='r', linewidth=2)
ax.scatter([1, 2, 3, 4], [1, 4, 9, 16], color='b', marker='o')
ax.set_xlim(0.5, 4.5)
```

```python
fig3 = plt.figure()
axx1 = fig3.add_subplot(211)
axx2 = fig3.add_subplot(212)
x = np.linspace(0, 10, 100)
y1, y2 = np.sin(x), np.cos(x)
axx1.plot(x, y1)
axx1.set_title('Signal 1')
axx2.plot(x, y2)
axx2.set_title('Signal 2')
```

```python
fig4, axes1 = plt.subplots(nrows=1, ncols=3, figsize=plt.figaspect(0.618))

x = np.linspace(0, 10, 100)

for i in range(6):
    axes1[0].plot(x, i * x)
    
for i, ls in enumerate(['-', '--', ':', '-.']):
    axes1[1].plot(x, np.cos(x) + i, linestyle=ls)
    
for i, (ls, mk) in enumerate(zip([' ', '-', ':'], ['o', '^', 's'])):
    axes1[2].plot(x, np.cos(x) + i * x, linestyle=ls, marker=mk, markevery=10)
    
```

```python
from mpl_toolkits import mplot3d
```

```python
fig5 = plt.figure()
ax5 = plt.axes(projection="3d")

z_line = np.linspace(0, 15, 100)
x_line = np.cos(z_line)
y_line = np.sin(z_line)

ax5.plot3D(x_line, y_line, z_line, 'gray')

z_points = 15 * np.random.random(100)
x_points = np.cos(z_points) + 0.1 * np.random.random(100)
y_points = np.sin(z_points) + 0.1 * np.random.random(100)

ax5.scatter3D(x_points, y_points, z_points, c=z_points, cmap='jet')
```
